```python
def testcase(m1, b1, m2, b2, size, width, n):
    # m1 defines the slope of the horizontal line
    plt.axis('equal')
    plt.xlim(0, size)
    plt.ylim(0, size)

    lines(m1, b1, m2, b2, size, width)
    
```

```python
def lines(m1, b1, m2, b2, size, width):
    b1a = b1
    b1b = b1 - width
    
    b2a = b2
    b2b = b2 - width
    
    plt.plot([0, size], [b1a, m1 * size + b1a])
    plt.plot([0, size], [b1b, m1 * size + b1b])
    
    plt.plot([0, size], [b2a, m2 * size + b2a])
    plt.plot([0, size], [b2b, m2 * size + b2b])
    plt.show()
```

```python
testcase(0.3, 5, -1.5, 15, 22, 1.5, 2)
```
